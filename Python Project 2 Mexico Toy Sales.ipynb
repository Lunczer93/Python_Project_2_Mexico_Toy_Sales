##### Grouping Store City by Stock On Hand with Ranking

```python
StoreCityStockOnHand =dfMergedLeftStoredAndProductToInventory.groupby('Store City')['Stock On Hand'].agg('sum').to_frame()
StoreCityStockOnHandSorted = StoreCityStockOnHand.sort_values('Stock On Hand', ascending=False).reset_index()
StoreCityStockOnHandSorted
```

```python
def Ranking(x):
    if x < 500:
        return 'Low Stock'
    elif x < 1000:
        return 'Med Stock'
    elif x < 1500:
        return 'Med-High Stock'
    else:
        return 'High Stock'



StoreCityStockOnHandSorted["Ranking"] = StoreCityStockOnHandSorted["Stock On Hand"].apply(Ranking)
```

```python
StoreCityStockOnHandSorted.set_index('Store City')
```

##### Visualising Stock on Hand by Product Name, then Ranking it

```python
StockOnHandByProductName=dfMergedLeftStoredAndProductToInventory.groupby(['Product Name'])['Stock On Hand'].agg("sum").to_frame()
StockOnHandByProductNameSorted = StockOnHandByProductName.sort_values(by='Stock On Hand', ascending=False).reset_index()
StockOnHandByProductNameSorted
```

```python
x = StockOnHandByProductNameSorted['Product Name']
y = StockOnHandByProductNameSorted['Stock On Hand']


plt.bar(x, y)

plt.title('Stock On Hand by Product Name')
plt.xlabel('Product Name')
plt.ylabel('Quantity of Product Name')

plt.xticks(rotation=90)
plt.figure(figsize=(100, 50))



plt.show()

```

```python
def RankingProductNameStockOnHand(x):
    if x <  500:
        return 'Low Level of Product in Stock'
    elif x <  1000:
        return 'Mid Level of Product in Stock'
    elif x < 1500:
        return 'Upper-Mid Level of Product in Stock'
    elif x < 2000:
        return 'High Level Of Product in Stock'
    else:
        return 'Upper High Level of Product in Stock'
    
    
StockOnHandByProductNameSorted['Ranking'] = StockOnHandByProductNameSorted['Stock On Hand'].apply(RankingProductNameStockOnHand)
```
