##### Grouping Store City by Stock On Hand with Ranking

```python
StoreCityStockOnHand =dfMergedLeftStoredAndProductToInventory.groupby('Store City')['Stock On Hand'].agg('sum').to_frame()
StoreCityStockOnHandSorted = StoreCityStockOnHand.sort_values('Stock On Hand', ascending=False).reset_index()
StoreCityStockOnHandSorted
```

```python
def Ranking(x):
    if x < 500:
        return 'Low Stock'
    elif x < 1000:
        return 'Med Stock'
    elif x < 1500:
        return 'Med-High Stock'
    else:
        return 'High Stock'



StoreCityStockOnHandSorted["Ranking"] = StoreCityStockOnHandSorted["Stock On Hand"].apply(Ranking)
```
